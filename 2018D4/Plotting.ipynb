```python
#!/usr/bin/python
# list comprehension, max(a, key = ), lambda, *args/**kwargs

filepath = 'input.txt'
import math

from datetime import datetime
startTime = datetime.now()

def oneCameFirst(one, two):
    sumOne = int(one[1:5])*10**8 + int(one[6:8])*10**6 + int(one[9:11])*10**4 + int(one[12:14])*10**2 + int(one[15:17])
    sumTwo = int(two[1:5])*10**8 + int(two[6:8])*10**6 + int(two[9:11])*10**4 + int(two[12:14])*10**2 + int(two[15:17])
    return (sumOne < sumTwo)

def main() -> "result1, result2":
    nums = []
    result1 , result2 = 0, 0
    with open(filepath) as f:
        for line in f.readlines():
            nums.append(line)
        f.close()

    # sloppy implementation of merge sort
    a = nums
    currentCompare = 2 
    
    x = [0]*len(a)
    plot = []
    while(True): #each layer
        size = int(currentCompare/2)
        for num in range(math.ceil(len(a)/currentCompare)):
            first = a[num*currentCompare:num*currentCompare+size]
            if len(a)-num*currentCompare < currentCompare:
                second = a[num*currentCompare+size:]
            else:
                second = a[num*currentCompare+size:num*currentCompare+size*2]
            checked = [len(first), len(second)]
                
            for i in range(len(first) + len(second)):
                if size-checked[0] == size:
                    a[num*currentCompare+i] = second[len(second)-checked[1]]
                    checked[1] = checked[1]-1  
                elif size-checked[1] == size or oneCameFirst(first[len(first)-checked[0]], second[len(second)-checked[1]]):
                    a[num*currentCompare+i] = first[len(first)-checked[0]]
                    checked[0] = checked[0]-1  
                else:
                    a[num*currentCompare+i] = second[len(second)-checked[1]]
                    checked[1] = checked[1]-1
        if currentCompare > len(a):
            break
            
        currentCompare = currentCompare*2

    currentLine = 0
    timesheets = {} # 'ID':[0]*60 
    while(currentLine < len(a)):
        shiftId = a[currentLine][26:a[currentLine].index(' b')]
        currentLine += 1
        numSleeps = 0
        while(True):
            try:
                a[currentLine+numSleeps*2].index('falls asleep') # throws errors if end of shift or end of list
                startTime = a[currentLine+numSleeps*2][15:17]
                endTime = a[currentLine+numSleeps*2+1][15:17]
                
                if shiftId not in timesheets:
                    timesheets[shiftId] = [0]*60
                    
                for i in range(int(startTime),int(endTime)):
                    timesheets[shiftId][i] += 1
                
                numSleeps += 1
            except (ValueError, IndexError): # 'waking up'
                currentLine = currentLine+numSleeps*2
                break
        
    timeSleeping = {k: sum(v) for (k, v) in timesheets.items()}
    guard = max(timeSleeping, key=timeSleeping.get)
    time = timesheets[guard].index(max(timesheets[guard]))
    result1 = int(guard)*int(time)
        
    m = {k: max(v) for (k, v) in timesheets.items()}
    guard = max(m, key=m.get)
    time = timesheets[guard].index(max(timesheets[guard]))
    result2 = int(guard)*int(time)

    return result1, result2

if __name__ == '__main__':
    print(main(), str(datetime.now() - startTime)[:-3])
```
