```python
from datetime import datetime
startTime = datetime.now()

from itertools import combinations

filepath = 'input.txt'
#filepath = 'test.txt'

r1, r2 = 0, 0

directions = [(1, 1), (-1, 1), (-1, -1), (1, -1) ]
x, y = 0, 0
d = 0
way_x, way_y = 10, 1

with open(filepath, 'r') as f:
    data = [line.strip() for line in f.readlines()]
    
for i in data:
    move, l = i[0], int(i[1:])
    
    if move == 'N':
        way_y += l
    elif move == 'S':
        way_y -= l
    elif move == 'E':
        way_x += l
    elif move == 'W':
        way_x -= l
    elif move == 'L':
        xx = way_x - x
        yy = way_y - y
        
        if xx >= 0 and yy >= 0:
            quad = 0
        elif xx < 0 and yy >= 0:
            quad = 1
        elif xx < 0 and yy < 0:
            quad = 2
        elif xx >= 0 and yy < 0:
            quad = 3
        else:
            pass
        #print(quad)
        
        if l == 90:
            quad = (quad+1)%4
        elif l == 180:
            quad = (quad+2)%4
        elif l == 270:
            quad = (quad+3)%4

        else:
            print('ERROR L', d)
        
        
        diff_x = abs(way_x - x)
        diff_y = abs(way_y - y)
        
        if l == 90 or l == 270:
            temp = diff_x
            diff_x = diff_y
            diff_y = temp
        
        way_x = x + diff_x * directions[quad][0]
        way_y = y + diff_y * directions[quad][1]
        
        #print(way_x, way_y, quad, diff_x, diff_y)
            
    elif move == 'R':
        xx = way_x - x
        yy = way_y - y
        
        if xx >= 0 and yy >= 0:
            quad = 0
        elif xx < 0 and yy >= 0:
            quad = 1
        elif xx < 0 and yy < 0:
            quad = 2
        elif xx >= 0 and yy < 0:
            quad = 3
        else:
            pass
        #print(quad)
        
        if l == 90:
            quad = (quad-1)%4
        elif l == 180:
            quad = (quad-2)%4
        elif l == 270:
            quad = (quad-3)%4

        else:
            print('ERROR R', d)
        
        
        diff_x = abs(way_x - x)
        diff_y = abs(way_y - y)
        
        if l == 90 or l == 270:
            temp = diff_x
            diff_x = diff_y
            diff_y = temp
        
        way_x = x + diff_x * directions[quad][0]
        way_y = y + diff_y * directions[quad][1]
        
        #print(way_x, way_y, quad, diff_x, diff_y)
        
    elif move == 'F':
        diff_x = way_x - x
        diff_y = way_y - y
        x += l * diff_x
        y += l * diff_y
        
        way_x += l * diff_x
        way_y += l * diff_y
        
    else:
        print('ERROR')
    
    #print('STEP', x, y, d, way_x, way_y, abs(way_x - x), abs(way_y - y))
    
print(abs(x) + abs(y))

print(r1, r2)
print(str(datetime.now() - startTime)[:-3]) # 25918798 3340942
```
