```python
#!/usr/bin/python

filepath = 'input.txt'

from datetime import datetime
import numpy
startTime = datetime.now()

WIDTH, HEIGHT = 50, 6
FONT_WIDTH, FONT_HEIGHT = 5, 6

operations = []
with open(filepath) as f:
    for line in f.readlines():
        operations.append(line.strip())
    f.close()
    
def main():
    grid = numpy.zeros((HEIGHT, WIDTH))
    
    for operation in operations:
        command = operation[:operation.index(' ')] 
        
        if command == 'rect':
            A = int(operation[operation.index(' ')+1:operation.index('x')])
            B = int(operation[operation.index('x')+1:])
            
            grid[0:B, 0:A] = 1
            
        elif command == 'rotate':
            direction = operation[7]
            index = int(operation[operation.index('=')+1:operation.index(' by')])
            size = int(operation[operation.index('by ')+3:])
            
            if direction =='c': #down
                grid[:, index] = numpy.roll(grid[:, index], size)
            elif direction == 'r': #side
                grid[index, :] = numpy.roll(grid[index], size)
            else:
                print("ELF NOT HAPPY")
                        
        else:
            print("ELF NOT HAPPY")
    
    for x in range(WIDTH//FONT_WIDTH):
        for y in range(FONT_HEIGHT):
            print(['#'  if char == 1 else ' ' for char in grid[y, x*FONT_WIDTH:x*FONT_WIDTH+FONT_WIDTH]])
        print()
        
    unique, counts = numpy.unique(grid, return_counts=True)
    return counts[numpy.where(unique == 1)][0]
    
if __name__ == '__main__':
    print(main(), str(datetime.now() - startTime)[:-3]) # 116 UPOJFLBCEZ 0:00:00.005
```
