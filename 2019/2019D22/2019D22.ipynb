```python
from collections import deque

instructions = []
with open('input.txt', 'r') as f:
    for line in f.readlines():
        instructions.append(line.strip())

deck_size = 10007
deck = deque(range(deck_size))

print('made deck')

def cut(deck, pos):
    deck.rotate(-1* int(pos))
    return deck

def increment(deck, skip):
    skip = int(skip)
    new_deck = deque([0] * len(deck))
    pos = 0
    n = 0
    
    while n < len(deck):
        new_deck[pos%len(deck)] = deck[n]
        pos += skip
        n += 1
    
    return new_deck

def stack(deck):
    deck.reverse()
    return deck

print(len(instructions))
#instructions = ['deal with increment 3']
for n, instruction in enumerate(instructions):
    if 'cut' in instruction:
        deck = cut(deck, instruction.split()[1])
    elif 'increment' in instruction:
        deck = increment(deck, instruction.split()[3])
    elif 'stack' in instruction:
        deck = stack(deck)
        
    print(n)
print(deck.index(2019))
```
