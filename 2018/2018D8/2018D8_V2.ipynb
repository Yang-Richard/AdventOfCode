```python
#!/usr/bin/python
filepath = 'input.txt'

from datetime import datetime
import re
startTime = datetime.now()

class Node(object):
    def __init__(self, name = '', parent = None, num_children = 0, num_metadata = 0):
        self.name = name
        self.parent = parent
        self.num_children = num_children
        self.num_metadata = num_metadata
        self.sum_metadata = 0
        
        self.children = []
        self.metadata = []

    def add_child(self, child):
        self.children.append(child)
        
    def add_metadata(self, metadata):
        for data in metadata:  
            self.metadata.append(data)
            self.sum_metadata += data

def main():
    result1, result2 = 0, 0
    
    with open(filepath) as f:
        for line in f.readlines():
            data = [int(i) for i in line.strip().split(' ')]
        f.close()

    tree = []
    current_parent = None
    
    i = 0
    while i < len(data):
        current_node = Node('', current_parent, data[i], data[i+1])
        tree.append(current_node)
        
        if current_parent != None:
            current_parent.add_child(current_node)
        
        if current_node.num_children == 0:
            metadata = data[i+2:i+2+current_node.num_metadata]
            current_node.add_metadata(metadata)
            i += current_node.num_metadata
            
            while current_node.parent != None:
                if current_node.parent.num_children == len(current_node.parent.children):
                    metadata = data[i+2:i+2+current_node.parent.num_metadata]
                    current_node.parent.add_metadata(metadata)
                    i += current_node.parent.num_metadata
                    
                    current_node = current_node.parent
                else:
                    break 
                    
            current_parent = current_node.parent
            
        else:        
            current_parent = current_node

        i += 2

        
    for current_node in tree:
        for metadata in current_node.metadata:
            result1 += metadata
            
        if current_node.num_children == 0:
            current_total_metadata = current_node.sum_metadata
                            
            while current_node.parent != None:
                index = current_node.parent.children.index(current_node)+1
                
                if index in current_node.parent.metadata:
                    multiplier = current_node.parent.metadata.count(index)
                    current_total_metadata *= multiplier
                else:
                    break
                
                current_node = current_node.parent
                
            if current_node.parent == None:
                result2 += current_total_metadata
                    
    return result1, result2
        
if __name__ == '__main__':
    print(main(), str(datetime.now() - startTime)[:-3]) # (42196, 33649) 0:00:00.012
```
